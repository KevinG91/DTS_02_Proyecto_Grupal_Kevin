```python
def get_coordinates(address):
    city = 'New York, USA'
    geocode_result = gmaps.geocode(str(address) +' '+ city)
    if len(geocode_result) > 0:
        return list(geocode_result[0]['geometry']['location'].values())
    else:
        return [np.NaN, np.NaN]
```

```python
coordinates = df1['Roadway_Name'].apply(lambda x: pd.Series(get_coordinates(x), index=['LATITUDE', 'LONGITUDE']))
```

```python
def plot_coordinates(lat,lon):
    import matplotlib.pyplot as plt
    from mpl_toolkits.basemap import Basemap
    
    # determinar el rango a imprimir usando min y max de latitud y longitud
    margin = 2 # Rango de separacion
    lat_min = min(df1['LATITUDE']) - margin
    lat_max = max (df1['LATITUDE']) + margin
    lon_min = min(df1['LONGITUDE']) - margin
    lon_max = max(df1['LONGITUDE']) + margin
    
    


    # crear un mapa usando BASEMAP
    m = Basemap(llcrnrlon=lon_min,
                llcrnrlat=lat_min,
                urcrnrlon=lon_max,
                urcrnrlat=lat_max,
                lat_0=(lat_max - lat_min)/2,
                lon_0=(lon_max-lon_min)/2,
                projection='merc',
                resolution = 'h',
                area_thresh=10000.,
                )
    m.drawcoastlines()
    m.drawcountries()
    m.drawstates()
    m.drawmapboundary(fill_color='#46bcec')
    m.fillcontinents(color = 'white',lake_color='#46bcec')
    # convertir latitud y longitud a un mapa de proyeccion con las coordenadas
    lons, lats = m((df1['LONGITUDE']), (df1['LATITUDE']))
    # plot points as red dots
    m.scatter(lons, lats, marker = 'o', color='r', zorder=5)
    plt.show()

plot_coordinates(df1['LATITUDE'].values, df1['LONGITUDE'].values)
```
